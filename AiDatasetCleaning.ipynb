```python
def get_job_category(job_title):
    if job_title in ['Cybersecurity Analyst', 'AI Researcher', 'UX Designer', 'Software Engineer', 'Data Scientist']:
        return 'Technology & Engineering'
    elif job_title in ['Sales Manager', 'HR Manager', 'Product Manager', 'Operations Manager']:
        return 'Management'
    elif job_title == 'Marketing Specialist':
        return 'Marketing'
    else:
        return 'Unknown'  

df['Job_Category'] = df['Job_Title'].apply(get_job_category)
```

```python
def categorize_by_region(Location):
    if Location == 'Dubai':
        return 'Middle East'
    elif Location in ['Singapore', 'Tokyo']:
        return 'Asia'
    elif Location in ['Berlin', 'London', 'Paris', 'Sydney', 'Toronto']:
        return 'Europe'
    elif Location in ['New York', 'Toronto']:
        return 'North America'
    elif Location == 'Sydney':
        return 'Australia'
    else:
        return 'Unknown'

df['Region'] = df['Location'].apply(categorize_by_region)
```
