```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sales Transaction v.4a.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head(20)
```

```python
df = df.drop(df[df['TransactionNo'].str.contains('C')].index)
df
```

```python
df['Sales'] = df['Price']* df['Quantity']
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
missing_customer_no = df[df['CustomerNo'].isnull()]
print(missing_customer_no)
```

```python
df.dropna(inplace=True)
```

```python
df['TransactionNo'] = df['TransactionNo'].astype(int)
df['CustomerNo'] = df['CustomerNo'].astype(int)
print(df.dtypes)
```

```python
df['MonthYear'] = df['Date'].dt.strftime('%Y-%m')
monthlysales = df.groupby('MonthYear')['Sales'].sum().reset_index()

plt.figure(figsize=(10,6))
plt.plot(monthlysales['MonthYear'], monthlysales['Sales'], marker='o', color='b')
plt.title('Sales Trend Over The Months')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.xticks(rotation = 45)
plt.ticklabel_format(style ='plain', axis ='y')
plt.grid(True)
plt.show()
```

```python
product_purchased = df.groupby('ProductName')['Quantity'].sum().reset_index().sort_values(by='Quantity', ascending=False)

plt.figure(figsize=(10,6))
top_10_products = product_purchased.head(10)
plt.bar(top_10_products['ProductName'], top_10_products['Quantity'])
plt.title('Top 10 Most Frequently Purchased Products')
plt.xlabel('ProductName')
plt.ylabel('Quantity Purchased')
plt.xticks(rotation=80)
plt.show() 
```

```python
customer_sales = df.groupby('CustomerNo')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)

plt.figure(figsize=(10,6))
top_10_customers = customer_sales.head(10)
plt.bar(top_10_customers['CustomerNo'].astype(str), top_10_customers['Sales'], color='purple')
plt.title('Top 10 Most Profitable Customers')
plt.xlabel('CustomerNo')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.ticklabel_format(style ='plain', axis ='y')
plt.tight_layout()
plt.show() 
```

```python
from IPython.display import FileLink
df.to_csv('Sales_Transactions_Cleaned.csv', index=False)
FileLink('Sales_Transactions_Cleaned.csv')
```

```python
import seaborn as sns 
transaction_summary = df.groupby('TransactionNo')['Quantity'].sum().reset_index().sort_values(by='Quantity', ascending=False)

plt.figure(figsize=(12, 6))
top_10 = transaction_summary.head(10)
sns.barplot(x='TransactionNo', y='Quantity', data=top_10, order=top_10['TransactionNo'], palette='BuPu')
plt.title("Top 10 Transactions by Total Quantity Purchased")
plt.xlabel('Transaction Number')
plt.ylabel('Total Quantity Purchased')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```
